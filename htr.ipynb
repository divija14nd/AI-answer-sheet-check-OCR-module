```python
def noise_removal(image):
    import numpy as np
    kernel = np.ones((2,2), np.uint8)
    image = cv2.erode(image, kernel, iterations=1)
    return image
```

```python
def threshold(image):
    """Apply thresholding to a grayscale image."""
    _, im_bw = cv2.threshold(image, 130, 255, cv2.THRESH_BINARY)
    return im_bw
```

```python
def sharpen(image):
    """Sharpen the image to enhance edges."""
    kernel = np.array([[0, -1, 0],
                       [-1, 5,-1],
                       [0, -1, 0]])
    return cv2.filter2D(image, -1, kernel)
```

```python
import pytesseract
```

```python
def ocr(image):
    ocr_res = pytesseract.image_to_string(image)
    return ocr_res

```

```python
def detect_horizontal_lines(image):
    """Detect horizontal lines in the image."""
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 150, apertureSize=3)
    lines = cv2.HoughLines(edges, 1, np.pi / 180, 200)
    
    if lines is not None:
        # Extract horizontal lines
        horizontal_lines = [line[0] for line in lines if np.abs(np.sin(line[0][1])) < 0.1]
        return sorted(horizontal_lines, key=lambda x: x[1])  # Sort by y-coordinate
    return []
```

```python


def get_ground_truth(image_path):
    """Extract ground truth text from the image."""
    img = cv2.imread(str(image_path))
    lines = detect_horizontal_lines(img)
    
    if len(lines) >= 2:
        y1 = int(lines[0][0][1])  
        y2 = int(lines[1][0][1])  
        
        # Crop the image between these lines
        height, width, _ = img.shape
        cropped_image = img[y1:y2, 0:width]
        
        ground_truth_text = ocr(cropped_image)
        return ground_truth_text.strip()
    
    return "Ground truth not available"

```

```python
from difflib import SequenceMatcher

def calculate_accuracy(ocr_output, ground_truth):
    """Calculate the accuracy of the OCR output compared to the ground truth."""
    return SequenceMatcher(None, ocr_output, ground_truth).ratio()
```

```python
def resize_image(image, scale_percent):
    """Resize image by a percentage."""
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)
    resized_image = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
    return resized_image
```

```python
if folder_path.is_dir():
    
    img_file = next(folder_path.glob('*.*'))  
    
    img = cv2.imread(str(img_file))

    scale_percent = 150
    resized_img = resize_image(img, scale_percent)
    gray_scaled_image = grayscale(resized_img)
    thresholded_image = threshold(gray_scaled_image)
    noise_removed_image = noise_removal(thresholded_image)

    sharpened_image = sharpen(noise_removed_image)
    original_ocr_output = ocr(img)
    processed_ocr_output = ocr(sharpened_image)
    ground_truth = get_ground_truth(img_file)

    original_accuracy = calculate_accuracy(original_ocr_output, ground_truth)
    processed_accuracy = calculate_accuracy(processed_ocr_output, ground_truth)

    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')  

    plt.subplot(1, 2, 2)
    plt.imshow(sharpened_image, cmap='gray')
    plt.title('Processed Image')
    plt.axis('off')  

    plt.show()

    print("Original Image OCR Output:")
    print(original_ocr_output)
    print("Processed Image OCR Output:")
    print(processed_ocr_output)
    print("Ground Truth:")
    print(ground_truth)
    print(f"Original Image OCR Accuracy: {original_accuracy:.2f}")
    print(f"Processed Image OCR Accuracy: {processed_accuracy:.2f}")

else:
    print(f"Error: The directory '{folder_path}' does not exist.")
```
